```python

#-----------------------------------------------------------
# Utilizando a API Yahooquery para pegar dados do Yahoo
#-----------------------------------------------------------


from yahooquery import Ticker
import csv
#-----------------------------------------------------------
# Utilizando o mplfinance para gerar o gráfico, pois ele contem 
#algumas funções que ajudam a exibir o gráfico de maneira mais fácil
#-----------------------------------------------------------
import mplfinance as mpf
import matplotlib.pyplot as plt
import pandas as pd
#-----------------------------------------------------------
# 3 ações escolhidas:
#BPAC11, (BTG PACTUAL), ITUB4 (ITAÚ), BBDC4 (BRADESCO).
#-----------------------------------------------------------
acao = Ticker("BPAC11.SA")
#acao = Ticker("ITUB4.SA")
#acao = Ticker("BBDC4.SA")

#-----------------------------------------------------------
#Testes com base da ação de 2019(antes da pandemia) e Base da ação de 2021(durante a pandemia)
#-----------------------------------------------------------
''' Extraindo periodo de X dias com intervalo de y minutos'''
#df = acao.history(start="2022-01-01", end="2021-12-31", interval = "30m" )

''' Extraindo periodo de x dias com intervalo de y minutos'''
df = acao.history(period="7d",  interval = "1m")


#-----------------------------------------------------------
#Transformando em CSV: https://www.acervolima.com.br/2020/08/criando-arquivo-csv-com-pandas.html
#-----------------------------------------------------------
df.to_csv('out.csv') 
#print(df)
df.head()
```

```python
#Dropei um dos Indexes do MultiIndex
# https://pandas.pydata.org/docs/reference/api/pandas.MultiIndex.droplevel.html 

df.index = df.index.droplevel('symbol')
```

```python
#-----------------------------------------------------------
#Transformando os dados em candlestick para melhorar a vizualização
#Fonte: https://python.plainenglish.io/a-simple-guide-to-plotly-for-plotting-financial-chart-54986c996682
#-----------------------------------------------------------


import plotly.graph_objects as go
import pandas as pd

import plotly.graph_objects as go
fig = go.Figure(go.Candlestick(x=df.index,
  open=df['open'],
  high=df['high'],
  low=df['low'],
  close=df['close']))


fig.update_xaxes(
        rangeslider_visible=True,
        rangebreaks=[
            # NOTE: Below values are bound (not single values), ie. hide x to y
            dict(bounds=["sat", "mon"]),  # hide weekends, eg. hide sat to before mon
            dict(bounds=[16, 10], pattern="hour"),  # hide hours outside of 9.30am-4pm
            # dict(values=["2019-12-25", "2020-12-24"])  # hide holidays (Christmas and New Year's, etc)
        ]
)
fig.update_layout(
        title='Stock Analysis',
        yaxis_title='Stock'
)

fig.show()
```
